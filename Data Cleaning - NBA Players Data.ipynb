```python
# Import the Pandas library
import pandas as pd
```

```python
# Import the dataframe from a .csv file
df = pd.read_csv('player_data.csv')
df
```

# ---------------------------------------------------------------------------------------------------------------

```python
# Corrects the titles names
df.columns = df.columns.str.title().str.replace('_',' ')
df
```

```python
# Normalize the positions into guards (G), fowards (F) and centers (C)
df = df.replace({'F-C':'C','C-F':'F','G-F':'G','F-G':'F'})
df
```

```python
# Converts feet-inches to meters
df[['feet', 'inches']] = df['Height'].str.split('-', expand=True)
df['feet'] = df['feet'].astype(float)
df['inches'] = df['inches'].astype(float)
df['Height'] = round((df['feet'] * 0.3048) + (df['inches'] * 0.0254),2)
df = df.drop(['feet','inches'],axis=1)
df
```

```python
# Converts pounds to kilograms
df['Weight'] = round(df['Weight']*0.453592,2)
df
```

```python
# Tuns the college column NaN into international
df['College'] = df['College'].fillna('International')
df
```

```python
# Penny Early was a woman who played just a second in the NBA
df = df.drop(df[df['Name']=='Penny Early'].index[0])
```

```python
# How many NaN is in each df column
df.isna().sum()
```

```python
# Checks if there is any NaN value in df
df[df.isna().any(axis=1)]
```

```python
# Excludes the lines that contain NaN values
df = df.dropna()
```

```python
# Split the birth date into birth day, month and year
df.insert(7, 'Birth Year', pd.to_datetime(df['Birth Date']).dt.year)
df.insert(7, 'Birth Month', pd.to_datetime(df['Birth Date']).dt.month)
df.insert(7, 'Birth Day', pd.to_datetime(df['Birth Date']).dt.day)
df = df.drop(['Birth Date'],axis=1)
df['Birth Year'] = df['Birth Year'].astype(int)
df['Birth Month'] = df['Birth Month'].astype(int)
df['Birth Day'] = df['Birth Day'].astype(int)
df
```

```python
# How many NaN is in each df column
df.isnull().sum()
```

```python
# df statistical resume
df.describe()
```

```python
# Cleaned df
df
```

```python
# Export .csv file to be used in other projects
df.to_csv('players_data_cleaned.csv', index=False)
```
