```python
# Imports the libraries
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
# Imports the data from a .csv file to a dataframe
df = pd.DataFrame()

df_patient = pd.read_csv(f"1.csv").drop('0', axis=1)
df_patient = df_patient.rename(columns={df_patient.columns[0]: 'X',df_patient.columns[1]: 'Y', df_patient.columns[2]: 'Z', df_patient.columns[3]: 'Label'})
df = pd.concat([df, df_patient], ignore_index=True)

df
```

```python
# Checks if there is any NaN
df.isna().sum()
```

```python
# Splits the dataframe into X and Y
df_x = df[['X','Y','Z']]
df_y = df['Label']
```

```python
# Scales the data
scaler = StandardScaler()
df_x = scaler.fit_transform(df_x)
```

```python
# Splits the data into train and test
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3,random_state=42)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Creates the SVC model
SVC_model = SVC()

# Trains the model
history = SVC_model.fit(x_train, y_train)

# Predicts the x_test data
y_pred = SVC_model.predict(x_test)

# Calculates the model accuracy
SVC_accuracy = accuracy_score(y_test, y_pred)
SVC_accuracy
```

```python
# Creates the MLP model
MLP_model = Sequential([
    Dense(x_train.shape[1], kernel_initializer='random_uniform', activation='relu', input_dim=x_train.shape[1]),
    Dense(64, kernel_initializer='random_uniform', activation='relu'),
    Dropout(0.1),
    Dense(64, kernel_initializer='random_uniform', activation='relu'),
    Dense(8, kernel_initializer='random_uniform', activation='softmax')])

MLP_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

# Trains the model
history = MLP_model.fit(x_train, y_train, validation_split=0.1, epochs=15)

# Calculate the model loss and accuracy
MLP_loss, MLP_accuracy = MLP_model.evaluate(x_test, y_test)
```

```python
# Displays and compares the models accuracy
df = pd.DataFrame({'Model':['MLP','SVC'],
                   'Accuracy':[round(MLP_accuracy,4),round(SVC_accuracy,4)]})

df
```
