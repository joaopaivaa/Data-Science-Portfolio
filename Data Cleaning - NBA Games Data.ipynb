```python
# Import the Pandas library
import pandas as pd
```

```python
# Import the dataframe from a .csv file
df_games = pd.read_csv('games_details.csv')
df_games
```

# ---------------------------------------------------------------------------------------------------------------

```python
# Displays the dataframe columns
df_games.columns
```

```python
# Drops specifc columns
df_games = df_games.drop(['GAME_ID','TEAM_ID','TEAM_CITY','PLAYER_ID','NICKNAME','COMMENT','START_POSITION','TEAM_ABBREVIATION'],axis=1)
df_games
```

```python
# How many NaN is in each df column
df_games.isna().sum()
```

```python
# Excludes the lines that contain NaN values
df_games = df_games.dropna()
df_games
```

```python
# Corrects the titles names
df_games.columns = df_games.columns.str.title().str.replace('_',' ')
df_games
```

```python
# Resets the dataframe index
df_games = df_games.reset_index().drop(['index'],axis=1)
```

```python
# Displays the final dataframe
df_games
```

```python
# Export .csv file to be used in other projects
df_games.to_csv('games_details_cleaned.csv', index=False)
```
